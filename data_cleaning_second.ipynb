```python
import pandas as pd
```

```python
data_import = pd.read_csv("data_cleaning_challenge.csv")
```

```python
data_import
```

```python
drop_extra = data_import.drop(columns=["Unnamed: 9", "Unnamed: 10"])
```

```python
drop_nans = drop_extra[drop_extra["Row Type"].notna()]
drop_nans
```

```python
columns_values = []
counter = 0

for i in drop_nans["Row Type"]:
    if "first name" in i:
        counter += 1
    columns_values.append(counter)
```

```python
iter_cols = drop_nans
iter_cols["Iterations"] = columns_values
```

```python
iter_cols
```

```python
drop_extra_rows = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_rows
```

```python
name_dataframe = drop_extra_rows[drop_extra_rows["Row Type"].str.contains("first name")]
```

```python
name_dataframe
```

```python
name_dataframe.drop(columns=["Speed1", "Speed2", "Electricity", "Effort", "Weight", "Torque"], inplace=True)
name_dataframe.rename(columns={"Row Type": "First Name", "Iter Number": "Last Name", "Power1": "Date"}, inplace=True)
name_dataframe
```

```python
name_dataframe["First Name"] = name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
name_dataframe
```

```python
noname_dataframe = drop_extra_rows[~drop_extra_rows["Row Type"].str.contains("first name")]
```

```python
noname_dataframe
```

```python
df = pd.merge(left=name_dataframe, right=noname_dataframe, how="inner", on="Iterations")
```

```python
df.to_csv('outputex2', index=False)
```
